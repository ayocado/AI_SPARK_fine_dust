##### 추가 및 가공한 칼럼
1. PM2.5_1 : 1시간 전 PM2.5
2. PM2.5_24 : 24시간 전 PM2.5
3. 2017, 2018, 2019, 2020, 2021 :
    - 2021이 test에만 있기 때문에 cat_features 적용 X
    - 따로 컬럼 생성
4. cat_features :
    - CatBoost는 cat_features로 칼럼을 지정하면, 학습할 때 알아서 범주화를 해줌
    - 적용하기 이전에 자료형을 object로 변경
    - cat_features = ['월', '일', '시', '측정소']
5. 최종 columns : 
    - ['PM2.5', '기온', '풍향', '풍속', '강수량', '습도', 'PM2.5_1', 'PM2.5_24', '2017', '2018', '2019', '2020', '2021']

```python
# 새로운 칼럼 추가를 위해 train_pm 복사
train_pm_shift = train_pm.copy()

# 1시간 전 PM2.5 칼럼 추가
train_pm_shift['PM2.5_1'] = train_pm_shift['PM2.5'].shift(periods=1, fill_value=0)

# 24시간 전 PM2.5 칼럼 추가
train_pm_shift['PM2.5_24'] = train_pm_shift['PM2.5'].shift(periods=24, fill_value=0)

# '년' 범주화
for i in range(2017, 2022):
    train_pm_shift[f'{i}'] = np.where(train_pm_shift['년']==i, 1, 0)
train_pm_shift.drop('년', axis=1, inplace=True)

# object 타입으로 변경
train_pm_shift['월'] = train_pm_shift['월'].astype('str')
train_pm_shift['일'] = train_pm_shift['일'].astype('str')
train_pm_shift['시'] = train_pm_shift['시'].astype('str')
train_pm_shift['측정소'] = train_pm_shift['측정소'].astype('str')

# object 변수 cat features 이용
cat_features = ['월', '일', '시', '측정소']
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import optuna 
from optuna import Trial, visualization
from optuna.samplers import TPESampler


x = train_pm_shift.drop('PM2.5', axis=1)
y = train_pm_shift.loc[:, 'PM2.5']

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=2023, stratify=train_pm_shift['측정소'])
```

```python
# random sampler
sampler = TPESampler(seed=10)

def objectiveCat(trial: Trial):
    params = {
        'learning_rate' : trial.suggest_loguniform('learning_rate', 0.001, 0.1),
        "depth": trial.suggest_int("depth", 1, 15),
        "iterations": trial.suggest_int("iteration", 1000, 10000),
        "l2_leaf_reg" : trial.suggest_int("l2_leaf_reg", 2, 30),
    }

    model = CatBoostRegressor(**params)
    model.fit(x_train, y_train, eval_set=[(x_val, y_val)], early_stopping_rounds=25, cat_features=cat_features, verbose=1)

    y_pred = model.predict(x_val)
    mse = mean_squared_error(y_val, y_pred)
    return mse

# MSE가 최소가 되는 방향으로 학습을 진행
studyCat = optuna.create_study(direction='minimize', sampler=sampler)

# n_trials 지정
studyCat.optimize(objectiveCat, n_trials=1)
```

```python
# optuna를 통한 best parameter를 이용하여, 전체 데이터로 best model 학습
model = CatBoostRegressor(learning_rate=0.034885205571560766, depth=1, iterations=6703, l2_leaf_reg=23)

model.fit(x, y)
```

# test data 전처리 및 model 적용
1. test_aws : 
    - KNN imputer로 PM2.5 제외 결측치 처리
    - 모든 AWS 병합
    - columns = ['기온', '풍향', '풍속', '강수량', '습도', '년', '월', '일', '시', '지점']
2. 사용한 방법 : 
    1. test_aws, pm_aws를 이용해서 PM 위치 별 기상값 계산
    2. 모델 적용을 위해 train_pm과 동일하게 전처리 진행
        - 년, 월, 일, 시, 측정소 칼럼 생성
        - '측정소'에 LabelEncoder
        - PM2.5_1, PM2.5_24, 2017, 2018, 2019, 2020, 2021 컬럼 생성
        - '월', '일', '시', '측정소' 타입 변경 -> object
        - cat_features = ['월', '일', '시', '측정소']
    3. for문으로 PM2.5 예측
        1. PM2.5가 결측치면,
        2. 해당 열을 input으로 설정
        3. input에 model 적용하여 PM2.5 예측
        4. 예측값을 i+1행의 PM2.5_1에 채워줌
        5. 만약 PM2.5_24가 결측치면,
        6. i-23행의 PM2.5를 PM2.5_24에 채워줌
        - PM 별로 적용

```python
# 시간 칼럼 생성
def datetime(data):
    for i in range(len(data)):
        data.loc[i, '년'] = 2021
        data.loc[i, '월'] = int(str(data.loc[i, '일시'])[:2])
        data.loc[i, '일'] = int(str(data.loc[i, '일시'])[3:5])
        data.loc[i, '시'] = int(str(data.loc[i, '일시'])[6:8])

    data.drop(['측정소', '연도', '일시'], axis=1, inplace=True)
    return data

# 칼럼 추가 함수
def new_columns(data):
    # 1시간 전 PM2.5 칼럼
    data['PM2.5_1'] = data['PM2.5'].shift(periods=1, fill_value=np.NaN)

    # 24시간 전 PM2.5 칼럼
    data['PM2.5_24'] = data['PM2.5'].shift(periods=24, fill_value=np.NaN)

    # '년' 범주화
    for i in range(2017, 2022):
        data[f'{i}'] = np.where(data['년']==i, 1, 0)
    data.drop('년', axis=1, inplace=True)

    # object 타입으로 변경
    data['월'] = data['월'].astype('str')
    data['일'] = data['일'].astype('str')
    data['시'] = data['시'].astype('str')
    data['측정소'] = data['측정소'].astype('str')
    return data

# object 변수 cat features 이용
cat_features = ['월', '일', '시', '측정소']

# PM 라벨링 딕셔너리
pm_dict = {'공주': 0, '노은동': 1, '논산': 2, '대천2동': 3, '독곶리': 4, '동문동': 5, '모종동': 6, '문창동': 7, '성성동': 8, '신방동': 9, '신흥동': 10,
           '아름동': 11, '예산군': 12, '읍내동': 13, '이원면': 14, '정림동': 15, '홍성읍': 16}
```
